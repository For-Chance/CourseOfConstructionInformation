### Task 1 - Loading data (2 points)
We will work with the data about world population and power consumption per capita. Please load the data and display the first 10 lines for each table.

Load the population data (`population.csv` file), and display the fist 10 rows. Remember that you have to import the necessary libraries.

```python
#ENTER YOUR CODE HERE
population = pd.read_csv('./population.csv')
population.head(10)
```

Now load the power consumption data (`power.csv` file), and display the fist 10 rows.

```python
#ENTER YOUR CODE HERE
power = pd.read_csv('./power.csv')
power.head(10)
```

### Task 2 - Merging and  data clearning (5 points)
Merge (join) the tables you loaded in task 1. Than change the column names, to make them easier to read. Use 'inner' join.



```python
# ENTER YOUR CODE HERE
merpp = pd.merge(power, population, how='inner', on='Country Code')
merpp.head(10)
```

```python
# ENTER YOUR CODE HERE
colnames = {
    "2000_x": "power_2000",
    "2005_x": "power_2005",
    "2010_x": "power_2010",
    "2000_y": "pop_2000",
    "2005_y": "pop_2005",
    "2010_y": "pop_2010"
}
merpp = merpp.rename(columns = colnames)
merpp.head(10)
```

### Task 3 - Selecting columns (5 points)
Create a new DataFrame (table), which contains only data for 2010. You should have the following columns:
- `Country Name` - name of the country
- `Country Code` - code of the country
- `pop_2010` - The population in 2010
- `power_2010` - The power consumption per capita in 2010

Then, create the column `power_total_2010`, which contains the overall usage of power for the year of 2010.

As a reminder:
- `power_2010` contains information about the usage of power per capita
- `pop_2010` cointains information aobut the population of the country


*Note: You may use different names for columns `pop_2010` and `power_2010` as long as they are intuitive.*

```python
#ENTER YOUR CODE HERE
tab_2010 = merpp[['Country Name', 'Country Code', 'pop_2010', 'power_2010']].copy()
tab_2010['power_total_2010'] = tab_2010['power_2010'] * tab_2010['pop_2010']
tab_2010.head(10)
```

### Task 4 - Filtering data (6 points)
Display all the rows, where total power usage is 5 times greater than the mean.

As a reminder:
- `power_2010` contains information about the usage of power per capita
- `pop_2010` cointains information aobut the population of the country


*Note 1: You may compute the mean using `df.power_total_2010.mean()`*

*Note 2: You must only display the results. You should not change your DataFrame, you will need it in the next task.*

```python
# ENTER YOUR CODE HERE
tab_2010_5times = tab_2010[tab_2010['power_total_2010'] > 5 * tab_2010['power_total_2010'].mean()]
tab_2010_5times.head()
```

### Task 5 - Grouping data (7 points)
<ul>
    <li>Load the `Countries-Continents.csv` table, and merge it with the table you prepared in **Task 3**.
    <li>Group your results by the name of the continent. For each continent compute:
    <ul>
        <li>The number of countries
        <li>The total usage of power (from `power_total_2010` column)
    </ul>
</ul>


```python
# ENTER YOUR CODE HERE
CC = pd.read_csv('Countries-Continents.csv')
merppc = pd.merge(CC, tab_2010, how='inner', left_on='Country', right_on='Country Name').drop('Country Name', axis=1)
merppc = merppc.groupby('Continent').agg({
    'Country':'count',
    'power_total_2010':'sum'
})
```
